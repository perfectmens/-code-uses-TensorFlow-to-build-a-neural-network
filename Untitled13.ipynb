```python
import numpy as np
import pandas as pd

```

```python
movie_info = pd.read_excel('/content/Movie Interests.xlsx')
```

```python
movie_info
```

```python
input_dataset = movie_info.drop(columns = ['Interest'])
input_dataset
```

```python
output_dataset = movie_info['Interest']
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
movie_model = DecisionTreeClassifier()
movie_model.fit(input_dataset,output_dataset )
```

```python
x=movie_model.predict([[90,0],[31,0]])
x
```
